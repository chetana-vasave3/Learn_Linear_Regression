# seaborn

- Seaborn is a data visuliztion library in python 
- It helps to create interesting an interactive plots to draw meaningful information from  data.
- Commonly used for DS and ML tasks.
- Backbone of EDA(Exploratory Data Analysis)
- The best use seaborn is that u will be able to get lot of statistical tools which will actually lep you  to understand  more about the data


- In datset you will be having Features like f1,f2,f3....
- based on the classification and regression problem you will be dividing your dataset.
- and this is with respect to supervised ML.
- Initailly u have to divide this features into Independent and dependent variable.
- If we have 2 features f1 and f2  we can draw 2D plot and by using this plot we will be able to find lot of info through this.
- Suppose we have 1 feature then this is basically called as univariate analysis.
- For 2 features this is called as bivariate analysis.
- now if we want to findout the relationship bw this two features and 
- this kind of univariate and bivariate analysis can be easily done with seaborn.
- It is also help us to analyze multiple features(multivariate analysis) in our DS.
- So for that we are going to use various kind of plots.



- Distplot   = univariate
- Joinplot   = bivariate analysis.
- Paireplot = multivariate analysis

### Dist plot

- It helps u to create histograms
- It represents the overall distribution of continuous data variables
- Represents univariate distribution of data i>e data distribution of a variable agains the density distribution
- Series, 1d-array, or list(only numeric values)

```python
a=np.random.rand(100)
sns.distplot(a)
```

```python
sns.distplot(data["tip"])
```

### join plot

- A join plot allows to study the relationship bw 2 numeric vaiables.
- It is usually a scatterplot,a hexbin plot a 2D historam or a 2D density plot
- used for bivariate anaysis

### Bivariate Analysis

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="hex") 

## data=ur dataset  
## kind="hex"- hexagonal shape relation
## we can give kind="scatter" | "kde" | "hist" | "hex" | "reg" | "resid" 
```

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="reg") 
```

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="scatter",color="g") 
```

```python
sns.jointplot(x="tip",y="total_bill",data=data,kind="kde") 
```

### Pair Plot

- used for multi variate anaysis
- A pair plot also known as scatterplot,in which one variable in the same data row is matched with another variable's value
- showing all variables paired with all the other variables

```python
sns.pairplot(data)
```

- using pairplot we can also findout relationship bw categorical data

```python
sns.pairplot(data,hue="sex")
```

```python
sns.pairplot(data,hue="smoker")
```
